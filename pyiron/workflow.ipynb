# Workflow notebook - introduction

This notebook gives the first insight into how to write a workflow in the pyiron frameework

```python
from pyiron_workflow.macro import Macro
from pyiron_workflow import Workflow
from pyiron_workflow.function import function_node, single_value_node
from pyiron_atomistics import Project
```

```python
@single_value_node()
def get_x(x_min, x_max, n):
    return np.linspace(x_min, x_max, n)

@single_value_node()
def get_y(x):
    result = x * 2
    return result

@single_value_node()
def get_z(x):
    result = x * 3
    return result
```

```python
wf = Workflow('MyWorkflow')
wf.add(get_x(x_min=0, x_max=1, n=1))
wf.add(get_y(x=wf.get_x), "y")
wf.add(get_z(x=wf.get_x), "z")
```

```python
wf.draw()
```
